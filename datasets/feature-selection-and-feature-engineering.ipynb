```python
import pandas as pd
import seaborn as sns

# Convert 'price' column to numeric if it's not already
train_df['price'] = pd.to_numeric(train_df['price'], errors='coerce')

# Drop non-numeric columns
numeric_df = train_df.select_dtypes(include=['number'])

# Drop rows with NaN values if any exist
numeric_df = numeric_df.dropna()

# Compute and sort correlation
correlation = numeric_df.corr()['price'].sort_values(ascending=False)
print(correlation)

# Plot heatmap (optional)
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
```
